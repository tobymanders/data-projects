Our next task is to construct a vocabulary from all of our reviews. We'll use a counter dictionary to count the frequency of each word in our dataset, and then we'll take the most frequent 10,000 words to build our vocabulary. We will use the vocabulary to 'unk' our dataset -- i.e. replace every word not in our vocabulary with the word 'unk'.
We also want to convert our dataset to an array of numbers -- i.e. 'tokens' -- and we can do both of these tasks in the same function.

```python
# Get most popular words and unk the rest
alltext = ''.join(df['text'].values).split()
```

```python
len(alltext)
```

```python
vocab_size = 10000

counter = sorted(Counter(alltext).items(), 
                 key=lambda x: -x[1])[:vocab_size] # sort by frequency
words, _ = list(zip(*counter))
word_to_ID = dict(zip(words, range(len(words))))
```

```python
list(counter)[:5]
```

```python
def tokenize_text(x):
    tokenized = [word_to_ID[word] if word in word_to_ID else 
                                         word_to_ID['<unk>'] for word in x.split()]
    return tokenized
            

df['text'] = df['text'].apply(tokenize_text)
```

```python
sample = 'i really loved dfalkjf especially the introduction'
print(f"Unk ID: {word_to_ID['<unk>']}")
print(tokenize_text(sample))
```

```python
seq_length = 250

def pad_text(x):
    rem = len(x) - seq_length
    if rem < 0:
        pad = [word_to_ID['<unk>']]*abs(rem)
        x.extend(pad)
        return x
    else:
        return x[:seq_length]
    
df['text'] = df['text'].apply(pad_text)
```

```python
X = np.array(list(df['text']))
y = df['overall'].isin(['4', '5'])
np.mean(y)
```

```python
# Read data
[word_to_ID, X, y] = pickle.load(open('audible_data.pkl', 'rb'))
```

```python
y = y.astype(int)
vocab_size = len(word_to_ID)
```

```python
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.1, random_state=42)
```

```python
# # 93.1%
# model = tf.keras.Sequential([
#     tf.keras.layers.Embedding(vocab_size, 8),
#     tf.keras.layers.Bidirectional(tf.keras.layers.LSTM(8)),
#     tf.keras.layers.Dense(16, activation='relu'),
#     tf.keras.layers.Dense(1, activation='sigmoid')
# ])


# # 93.5%
# model = tf.keras.Sequential([
#     tf.keras.layers.Embedding(vocab_size, 128),
#     tf.keras.layers.Dropout(rate=0.1),
#     tf.keras.layers.Bidirectional(tf.keras.layers.LSTM(32)),
#     tf.keras.layers.Dense(64, activation='relu'),
#     tf.keras.layers.Dropout(rate=0.1),
#     tf.keras.layers.Dense(1, activation='sigmoid')
# ])

# # 93.68%
# model = tf.keras.Sequential([
#     tf.keras.layers.Embedding(vocab_size, 64),
#     tf.keras.layers.Bidirectional(tf.keras.layers.LSTM(
#         64, return_sequences=True)),
#     tf.keras.layers.Bidirectional(tf.keras.layers.LSTM(32)),
#     tf.keras.layers.Dense(64, activation='relu'),
#     tf.keras.layers.Dropout(rate=0.4),
#     tf.keras.layers.Dense(1, activation='sigmoid')
# ])

# 93.76%
model = tf.keras.Sequential([
    tf.keras.layers.Embedding(vocab_size, 128),
    tf.keras.layers.Conv1D(filters=32, kernel_size=3, padding='same', activation='relu'),
    tf.keras.layers.MaxPooling1D(pool_size=2),
    tf.keras.layers.Conv1D(filters=64, kernel_size=3, padding='same', activation='relu'),
    tf.keras.layers.MaxPooling1D(pool_size=2),
    tf.keras.layers.Bidirectional(tf.keras.layers.LSTM(64)),
    tf.keras.layers.Dense(128, activation='relu'),
    tf.keras.layers.Dropout(rate=0.4),
    tf.keras.layers.Dense(1, activation='sigmoid')
])
```

```python
model.compile(loss='binary_crossentropy',
              optimizer='adam',
              metrics=['accuracy'])
```

```python
model.summary()
```

```python
es = EarlyStopping(monitor='val_loss', mode='min')
```

```python
import time
start = time.time()
history = model.fit(X_train,
                    y_train,
                    epochs=50,
                    batch_size=2048,
                    validation_data=(X_test, y_test),
                    callbacks=[es])
end = time.time()
print((end-start)/60)
```

```python
results = model.evaluate(X_test, y_test, batch_size=2048)

print(results)
```

```python
import matplotlib.pyplot as plt

def plot_graphs(history, string):
    plt.plot(history.history[string])
    plt.plot(history.history['val_'+string])
    plt.xlabel("Epochs")
    plt.ylabel(string)
    plt.legend([string, 'val_'+string])
    plt.title(f'Epoch vs. {string}')
```

```python
plt.figure(figsize=(10,4))
plt.subplot(1,2,1)
plot_graphs(history, 'acc')

plt.subplot(1,2,2)
plot_graphs(history, 'loss')

save_fig('epochs_vs_accuracy')
```

```python
from sklearn.metrics import roc_curve
y_pred = model.predict(X_test[:10000, :], batch_size=5000).ravel()
fpr_keras, tpr_keras, thresholds_keras = roc_curve(y_test[:10000], y_pred)
```

```python
from sklearn.metrics import auc
auc_keras = auc(fpr_keras, tpr_keras)
```

```python
plt.style.use('seaborn')
plt.figure(figsize=(8,8))
plt.plot([0, 1], [0, 1], 'k--')
plt.plot(fpr_keras, tpr_keras, label='AUC = {:.3f}'.format(auc_keras))
plt.legend(loc='best')
plt.title('ROC Curve')
plt.xlabel('False positive rate')
plt.ylabel('True positive rate')
save_fig('roc_curve')
```
